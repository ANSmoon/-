```python
"""
author : 2018114657 문범수
file : Bigdata_HW2
object : 공공데이터 크롤링 및 파일 저장
date : 2022-10-12
"""

import pandas as pd
import requests
import urllib.request
import json
from urllib.request import urlopen

url = 'http://openapi.seoul.go.kr:8088/684252776b6d756e3936754a586449/json/ChunmanFreeSuggestions/1/1000/'

response = requests.get(url)

response = urllib.request.urlopen(url)
json_str = response.read().decode("utf-8")

json_object = json.loads(json_str)
json_object

df=pd.json_normalize(json_object['ChunmanFreeSuggestions']['row'])
df.to_csv("민주주의서울자유제안.csv",index = False)

data = pd.read_csv("민주주의서울자유제안.csv")
data
```

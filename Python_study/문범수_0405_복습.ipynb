```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

### 머신러닝에 학습시키기 위해서는 
- 필수적으로 수치형 data만 가능함.[int, float]

```python
df_sp= df.select_dtypes(include=[np.number])
```

```python
df_sp = df_sp.drop('age', axis=1)
```

### 머신러닝은 지도학습의 개념
- 지도학습이라면 정답이 필요
- 정답은 흔히 말하는 y, label 개념
- 타이타닉 data에서는 생존의 개념 survived column이 정답인 것
- 머신러닝에 학습시키기 위해서는 data 분류를 해야함
- 머신에게 학습시킬 data와 실제 test 할 data를 나눠서 학습시킨다.
- train, test로 data를 나눈다.

```python
y_df = df_sp['survived'] #y 값을 의미
```

```python
y_df
```

```python
x_df = df_sp.drop('survived',axis=1)
```

```python
x_df
```

### train, test를 나눠서 진행한다.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.3, random_state=111)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

### 머신러닝 알고리즘을 이용해 titani 생존자 예측
- dt
- rf
- lr
- 대표적인 분류 세가지 모델을 이용해 정확도를 비교해 보자!

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 3가지를 변수에 넣는다. 
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(random_state=111)
```

```python
dt_clf
```

```python
rf_clf
```

```python
lr_clf
```

### 해당 모델에 data를 넣고 정확도 확인

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도:{0:.4f}'.format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RF정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))


lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```

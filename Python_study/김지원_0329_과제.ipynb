## 필수과제2
- 반복문을 이용해서 spam 데이터의 text를 전처리 해주세요
- 모든 값들을 다 특수문자를 제거하고 전처리 해주시고
- 해당 데이터를 빈도를 분석해 주세요.
- 해당 text 데이터에 어떤 문자가 가장 많은 빈도를 보이는지 찾아주세요!
- Go 가장 많았다 - 몇 개가 있는지?

- 어떤 데이터가 가장 많은 분포를 가지고 있는지 확인해 주세요!

```python
import pandas as pd
import string
```

```python
string.punctuation
```

```python
spam = pd.read_csv('spam.csv')
```

```python
spam
```

```python
def no(texts):                  # 특수문자 제거 함수
 aa = []
 for i in texts:
    if i not in string.punctuation:
        aa.append(i)
 aa=''.join(aa)            # 해당 조건을 없이 실행시켰더니 comma가 제거되지않음 
 return aa
```

```python
spam['text'] = spam['text'].apply(no)
spam
```

```python
spam.text
```

```python
a = []
for text in spam.text:
  for word in text.split(" "):
    a.append(word)
```

```python
a
```

```python
# 구글링해보니 빈도 수를 셀 때 value_counts 함수를 사용하는 것 같다  
## 함수를 사용하기 전에 리스트를 데이터프레임으로 만들어야 세기 편할 것 같다
```

```python
b = pd.DataFrame(a,columns = ['text'])
b
```

```python
bb = b['text'].value_counts()
bb
```

```python
cc = pd.DataFrame(bb)
cc
```

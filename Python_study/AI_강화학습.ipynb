```python
def set_initial_state(rooms = 6): #ì´ˆê¸°ìƒíƒœì˜ ë£¸ ì„¤ì •
    return np.random.randint(0, rooms)


def get_action(current_state, reward_matrix): #ìƒíƒœì§€ì •í›„ ê°€ëŠ¥í•œ  Action ì„ íƒ
    valid_actions = []
    for action in enumerate(reward_matrix[current_state]):
        if action[1] != -1:
            valid_actions += [action[0]]
    return random.choice(valid_actions)

    
def take_action(current_state, reward_matrix, gamma, verbose = False): #í•˜ë‚˜ì˜ ì‘ì—… ìˆ˜í–‰
    action = get_action(current_state, reward_matrix)
    sa_reward = reward_matrix[current_state, action] # í˜„ì¬ ìƒíƒœ-í–‰ë™ ë³´ìƒê°’
    ns_reward = max(q_matrix[action,]) # ë‹¤ìŒ ìƒíƒœ-í–‰ë™ ë³´ìƒê°’
    q_current_state = sa_reward + (gamma * ns_reward)
    q_matrix[current_state, action]  = q_current_state # q-matrixì˜ ë³€ê²½
    new_state = action
    if verbose: 
        print(q_matrix)
        print(f"Old State: {current_state} | New State: {new_state}\n\n")
        if new_state == 5:
              print(f"Agent has reached it's goal!") 
    return new_state
    
def initialize_episode(reward_matrix, initial_state, gamma, verbose = False):
  # ì—ì´ì „íŠ¸ê°€ ëª©í‘œ ìƒíƒœê°€ ë ë•Œê¹Œì§€ ì‹œë®¬ë ˆì´ì…˜ì„ ìˆ˜í–‰
    current_state = initial_state
    while True: # 5ë²ˆë°©ìœ¼ë¡œ ì´ˆê¸°í™” ë˜ë©´ ì•ˆë˜ë¯€ë¡œ 5ë²ˆë°©ì€ ì„ íƒí•˜ì§€ ì•ŠìŒ
        current_state = take_action(current_state, reward_matrix, gamma, verbose)
        if current_state == 5:
            break
        
def train_agent(iterations, reward_matrix, gamma, verbose = False):
  # ì§€ì •ëœ ìˆ˜ì˜ ì—í”¼ì†Œë“œë¥¼ ì‹¤í–‰í•œ í›„ í–‰ë ¬ì„ ì •ê·œí™”
    print("Training in progress..")
    for episode in range(iterations):
        initial_state = set_initial_state()
        initialize_episode(reward_matrix, initial_state, gamma, verbose)
    print("Training complete! ğŸ’¯ğŸ’¯ğŸ’¯ğŸ’¯")
    return q_matrix
   
def normalize_matrix(q_matrix):
    normalized_q = q_matrix / max(q_matrix[q_matrix.nonzero()]) * 100
    return normalized_q.astype(int)
```

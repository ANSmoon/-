```python
import missingno as msno
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("bike_sharing_daily.csv")
```

```python
df
```

```python
msno.matrix(df)
plt.show()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
#temp data를 사용하기
mean=df['temp'].mean()
mean
```

### ffill : 누락값이 나타나기 이전의 값으로 누락값을 변경

```python
df['temp'].fillna(mean).to_frame()
```

```python
####기존 data 유지하면서 0으로 대체 크게 의미 X
df['temp'].fillna(0).to_frame()
```

```python
df['temp'].fillna(method='ffill').to_frame()
```

```python
df.temp.to_frame()
```

```python
df['temp'].fillna(method='bfill').to_frame()
```

### Interpolate

```python
df['temp'].interpolate().to_frame()
```

```python
bike=df.iloc[0:20, 9:]
bike
```

```python
bike['temp'].fillna(method='bfill').to_frame()
```

```python
bike['temp'].plot() #결측값 존재
```

```python
bike['temp'].fillna(0).plot()
```

```python
bike['temp'].fillna(mean).plot()
```

```python
mean
```

### 선형 보간법[다양한 차수가 존재하니 확인하면서 할 것]

```python
bike['temp'].interpolate(method='linear').plot()
```

```python
# 항 차수 증가
bike['temp'].interpolate(method='slinear').plot()
```

```python
bike['temp'].interpolate(method='cubic').plot()
```

```python
# 2차항(spline)
bike['temp'].interpolate(method='spline', order=2).plot()
```

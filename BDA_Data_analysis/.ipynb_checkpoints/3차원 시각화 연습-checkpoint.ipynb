## 3D Plot Example

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 3차원 함수 예제 => x^2 + y^2 를 의미
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
# 데이터 생성
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = f(x, y)
```

```python
# 3D 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
# 킹갓 viridis

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot Example')

plt.show()
```

viridis는 matplotlib에서 제공하는 색상 맵 중 하나로, 시각화에서 데이터 값을 효과적으로 전달하기 위한 색상 척도입니다. 이 색상 맵은 특히 연속적인 데이터의 시각화에 적합하며, 인지적으로 일관성 있고 친화적인 색상을 사용합니다.

viridis는 기본적으로 matplotlib 2.0 버전 이후의 버전에서 색상 맵으로 도입되었습니다. 이전에는 jet이나 rainbow 등의 색상 맵이 주로 사용되었지만, 이러한 색상 맵들은 일부 문제가 있어서 viridis와 같은 향상된 색상 맵이 권장되고 있습니다.

## 3D Scatter Plot Example

```python
# 랜덤한 데이터 생성
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)
```

```python
# 3D 산점도 플로팅 (색상은 z 값에 따라 다름)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot with Color Based on Z')

# colorbar 추가
fig.colorbar(scatter, ax=ax, pad=0.1)

plt.show()
```

- 왜 한글 안됨

- scatter 함수를 사용하면 3D 공간에서 점들을 효과적으로 시각화할 수 있습니다. 아래는 scatter 함수를 사용하여 3D 산점도를 만드는 간단한 예제 코드입니다. 이 예제에서는 NumPy를 사용하여 무작위로 생성된 데이터를 사용합니다.

- 만약 색상을 변수에 따라 다르게 표현하고 싶다면, c 인자에 배열을 전달하면 됩니다. 

## 3D Line Plotting

```python
# 데이터 생성
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1

x = r * np.sin(theta)
y = r * np.cos(theta)
```

```python
# 3D 선 그래프 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D Spiral')

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Line Plot Example')
ax.legend()

plt.show()
```

## 3D Contour Plot

```python
# 함수 정의
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
# 데이터 생성
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = f(x, y)
```

```python
# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Contour Plot Example')

plt.show()
```

```python
# 3D 등고선 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(x, y, z, 50, cmap='viridis')
```

- 등고선 플로팅
- 
linspace는 NumPy 라이브러리에서 제공하는 함수 중 하나로, 주어진 범위에서 균일하게 간격을 둔 수열을 생성하는 데 사용됩니다. 일반적으로 numpy.linspace(start, stop, num) 형태로 사용되며, 이 함수는 start에서 stop 사이의 값을 num 개로 나누어 갖는 배열을 반환합니다.

- 여기서 각 인자의 의미는 다음과 같습니다:

- start: 시작 값
- stop: 종료 값
- num: 생성할 배열의 요소 수
- 예를 들어, np.linspace(0, 1, 5)는 0부터 1까지의 범위에서 5개의 요소를 가진 배열을 생성합니다. 결과적으로 배열은 [0. , 0.25, 0.5 , 0.75, 1. ]와 같이 값이 균일하게 분포됩니다.

## 3D Surface & Contour Plots

```python
# 함수 정의
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) * np.exp(-0.1*(x**2 + y**2))
```

```python
# 데이터 생성
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = f(x, y)

# 3D 서피스와 등고선 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)
contour = ax.contour(x, y, z, zdir='z', offset=np.min(z), cmap='viridis', linestyles='dashed', alpha=0.5)

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('Combination of 3D Surface and Contour Plots')

# colorbar 추가
fig.colorbar(surface, ax=ax, pad=0.1)

plt.show()
```

- 혼종스 탄생
- 등고선과 subplot 함께 사용

## Meshgrid 활용

```python
# 함수 정의
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# 데이터 생성
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)   # => mesh 추가 부분
z = f(x, y)

# 3D 서피스 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Surface Plot with Meshgrid')

# colorbar 추가
fig.colorbar(surface, ax=ax, pad=0.1)

plt.show()
```

## Parametric Surface

```python
# 매개변수 방정식 정의
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

# 매개변수 방정식으로부터 x, y, z 계산
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

# 세상 미친 방정식

# 3D 서피스 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('Parametric Surface Plot')

# colorbar 추가
fig.colorbar(surface, ax=ax, pad=0.1)

plt.show()
```

## 3D Box plot

```python
# 무작위 데이터 생성
np.random.seed(42)
data = np.random.randn(100, 3)

# 3D 박스 플로팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.boxplot(data)

# 플롯 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Box Plot Example')

plt.show()
```

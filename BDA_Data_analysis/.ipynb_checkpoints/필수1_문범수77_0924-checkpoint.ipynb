- 다양하게 통계치를 지정할 수 있다. ( 평균, 최빈값이 등등 ) 결과가 어떤식으로 바뀌는지 mpg 데이터를 가지고 확인해 주세요!

```python
# explicitly require this experimental feature
from sklearn.experimental import enable_iterative_imputer  # noqa

# now you can import normally from sklearn.impute
from sklearn.impute import IterativeImputer # 임퓨터 불러오기
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df=sns.load_dataset('mpg')
```

```python
df
```

```python
# 결측값이 Horsepower에 존재
df.isna().sum()
df.reset_index(inplace = True)
```

```python
#현재 Na값에 대한 clear
df.dropna(inplace = True)
```

```python
np.random.seed(230934)
na_index = np.random.choice(df.index, size=int(len(df.index)*0.3))
df_na = df.copy().drop(columns = 'index')
df_na.iloc[na_index,3] = np.nan # displacement 열에 결측값 존재
```

```python
df_na.isna().sum()
```

```python
df_na.columns
```

```python
column_list = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
       'acceleration', 'model_year']
df_na = df_na[column_list]
```

Mean 값 대체

```python
mean = SimpleImputer(strategy='mean')
df_mean = mean.fit_transform(df_na)
df_mean = pd.DataFrame(df_mean, columns = column_list)
```

```python
df_mean['displacement'].plot()
```

Median 대체

```python
median = SimpleImputer(strategy = 'median')
df_median = median.fit_transform(df_na)
df_median = pd.DataFrame(df_median, columns = column_list)
```

```python
df_median['displacement'].plot()
```

most frequent 대체

```python
mostfrq_imputer = SimpleImputer(strategy = 'most_frequent')
df_mostfrq = mostfrq_imputer.fit_transform(df_na)
df_mostfrq = pd.DataFrame(df_mostfrq, columns = column_list)
```

```python
df_mostfrq['displacement'].plot()
```

Constant 대체

```python
zero = SimpleImputer(strategy = 'constant', fill_value= np.min(df_na.displacement)) # 최소값으로 대체하자
df_zero = zero.fit_transform(df_na)
df_zero = pd.DataFrame(df_zero, columns = column_list)
```

```python
df_zero['displacement'].plot()
```

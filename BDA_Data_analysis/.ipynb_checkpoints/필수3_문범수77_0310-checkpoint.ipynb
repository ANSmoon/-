### 필수과제3- x feature 스케일링 진행
- 기존에 스케일링 진행하지 않은 피처들 스케일링

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv('trainby.csv')
train
```

```python
train_sp=train[['season','holiday','workingday','weather','temp','atemp','count']]
train_sp['count_log']=np.log(train_sp['count'])
train_sp
```

```python
train_fit_X=train_sp[['season','holiday','workingday','weather','temp','atemp']]
```

```python
train_fit_Y=train_sp['count_log']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_st=StandardScaler()
scaler_st.fit(train_fit_X)
train_fit_X_st=scaler_st.transform(train_fit_X)

train_fit_X_st=pd.DataFrame(data=train_fit_X_st, columns=train_fit_X.columns)
train_fit_X_st
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_fit_X_st, train_fit_Y, test_size=0.3, random_state=111)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
pred_train_1=model.predict(X_train)
pred_test_1=model.predict(X_test)
```

```python
mse_train_1=mean_squared_error(y_true=y_train, y_pred=pred_train_1)
mse_test_1=mean_squared_error(y_true=y_test, y_pred=pred_test_1)
```

```python
print(mse_train_1)
print(mse_test_1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_mm=MinMaxScaler()
scaler_mm.fit(train_fit_X)
train_fit_X_mm=scaler_mm.transform(train_fit_X)

train_fit_X_mm=pd.DataFrame(data=train_fit_X_mm, columns=train_fit_X.columns)
train_fit_X_mm
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_fit_X_mm, train_fit_Y, test_size=0.3, random_state=111)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
pred_train_2=model.predict(X_train)
pred_test_2=model.predict(X_test)
```

```python
mse_train_2=mean_squared_error(y_true=y_train, y_pred=pred_train_2)
mse_test_2=mean_squared_error(y_true=y_test, y_pred=pred_test_2)
```

```python
print(mse_train_2)
print(mse_test_2)
```

```python
mse_test_1==mse_test_2
```

### 파생변수란?
- 외부변수, 파생변수
- 통상적으로는 내부 data를 가지고 새로운 변수를 만드는 것
- 재료 조건이 내부 data 안에 있고, 그 data를 가지고 -> 새로운 변수를 만든다.

- 만드는 이유 
- 1. data 분석에 필요한 경우 만들 수 있다. 새로운 평균이라는 기준점을 가지고 분석을 해야 하는 경우(매출 CAGR 3개년치 성장율)
- 2. modeling을 통해 피처에 대한 엔지니어링을 진행하는 경우
- 기타 등등..

```python
df['3_mean'] = (df['math'] + df['english'] + df['science']) / 3
```

```python
df['3_mean']
```

```python
# assign 함수 이용해 파생변수 만들기
# assign 이용해서 만드는 것 뿐

df.assign(mean_t = (df['math'] + df['english'] + df['science'] )/ 3)
```

### groupby
- group화 시키기
- group으로 통계치 계산 or counting 계산 등등
- 결국은 group으로 무엇을 묶는다는 것이 중요 포인트

- df.groupby(기준).agg(변수, 요약통계치) 결과 확인

```python
df
# nclass 묶으면 반별 어떤 결과가 나올 것
# groupby는 agg 함수와 같이 많이 사용
df.groupby('nclass')
```

```python
df.groupby('nclass').agg
```

```python
df.groupby('nclass').agg(ncl_math = ('math', 'mean'))
# nclass 별로 수학 평균 삽입
```

```python
df.groupby('nclass').agg(ncl_math = ('math', 'sum'))
```

```python
df.groupby('nclass').agg(ncl_math = ('math', 'median'))
```

```python
df_gp = df.groupby('nclass').agg(ncl_math = ('math', 'std'),
                        ncl_eng = ('english', 'sum'),
                        ncl_math_mean = ('math', "mean"))
```

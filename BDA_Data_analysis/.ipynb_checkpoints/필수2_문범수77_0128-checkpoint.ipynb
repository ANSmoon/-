### 필수과제2 ( 패키지를 이용해서 tag 하여 진행 )
- 명사, 동사 정도만 추출하고
- 의미있는 단어들은 무엇인지 확인해 보기
- 해당 단어들과 label의 관계를 살펴보자!

```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
```

```python
k = Okt()
```

```python
df = pd.read_csv('movie_rv.csv')
df
```

```python
sample = df[0:100]
df1 = sample['document']
```

```python
df1
```

```python
combined_text = ' '.join(df1)  
```

```python
print(combined_text)
```

```python
k = Okt()
```

```python
n = k.nouns(combined_text)
print(n);
```

```python
pos = k.pos(combined_text)
v = [word for word, pos in pos if pos.startswith('Verb')]
print(v)
```

```python
from collections import Counter
```

```python
n_result = Counter(n)
sorted_result = sorted(n_result.items(), key=lambda x: x[1], reverse=True)
sorted_result # good : 최고, bad : 쓰레기
```

```python
v_result = Counter(v)
sorted_result = sorted(v_result.items(), key=lambda x: x[1], reverse=True)
sorted_result
```

### Groupby

```python
#그냥 바로 컬럼을 잡고 통계치를 볼 수 있다.
df.groupby('year').lifeExp.sum()
```

```python
df.groupby('year').lifeExp.mean()
```

```python
#2가지 이상을 잡게 된다면?
df.groupby('year')[['lifeExp','pop']].mean()
```

```python
#df.groupby('year').lifeExp.agg(내가 원하는 함수를 넣어서 사용할 수 있다.)
```

```python
#내가 원하는 함수를 만들어서 넣을 수 있다.
def my_mean(values):
    n = len(values)
    sum_1 = 0
    for value in values:
        sum_1+= value
    return sum_1/n
#내가 원하는 함수를 만들어서 agg 같이 넣을 수 있다.
```

```python
df.groupby('year').lifeExp.agg(my_mean)
```

```python
df.groupby('year').lifeExp.agg(my_mean)
```

- numpy를 이용해서 추가할 수 있다.
- agg 함수에 추가

```python
import numpy as np
df.groupby('year').lifeExp.agg([np.mean, np.std, np.count_nonzero])
```

### merge 
- merge 데이터 데이터를 합치는 경우에 발생한다.
- 실무에서는 DA sql 충분히 가능하지만,
- DS 데이터셋을 정리하는 경우도 있다. 
- 테이블을 병합시키고 정리해서 우리가 원하는 테스트셋을 만드는 경우 
- pandas 많이사용한다.

- 데이터간의 병합을 통해 내가원하는 새로운 테이블 만드는 개념
- merge
- concat

```python
test1 = pd.DataFrame({'class':['파문기','데분기','데분중','데분고'],
                    '인원':[100,200,300,40]})
test2 = pd.DataFrame({'class':['파문기','데분기','데분중','데분고'],
                    '벌점평균':[5,6,7,3]})
```

```python
print(test1)

print(test2)
```

- test1과 test2
- class에 따라 인원, 벌점평균을 같이 볼 수 있는 테이블 만드는 경우

- 공통의 키로 가지고 가서 class를 잡고 병합을 한다.

```python
# how 조인의 방법 = left join, right join inner join, outer join
# on 공통의 컬럼
pd.merge(test1, test2, how='left', on='class')
```

```python
##concat
## 덩어리 + 덩어리 
## 공통적인 키를 잡는 게 아니라 테이블 테이블, 덩어리 덩어리 두 개가 붙는 개념

sp_1=pd.concat([test1, test2],axis=0) ## 공통적인 컬럼을 지정하지 않는다.
```

```python
sp=pd.concat([test1, test2],axis=1) ## 공통적인 컬럼을 지정하지 않는다.
```

```python
sp
```

```python
sp_1.loc[1]
```

```python
sp_1.iloc[1]
```

- melt함수
- 깔끔하게 데이터를 가공하는 문법
---- 
- 기준을 잡는 설정 
- id_vars 위치를 그대로 유지할 열의 이름 지정
- value_vars 행으로 위치를 변경할 열의 이름
- var_name value_vars 위치로 변경한 열의 이름 지정
- value_name var_name 위치로 변경한 열의 이름 지정

```python
pd.melt(pew,id_vars='religion')
```

```python
pd.melt(pew,id_vars='religion',var_name='income',value_name='count')
```

```python
np.transpose(pew) #축의 반전의 개념
```

## 컬럼의 여러가지 의미가 있는 데이터의 경우 어떤 식으로 나누면 좋을까?
- 컬럼을 나눠서 새로운 컬럼을 만들 수 있다.
- 피처엔지니어링의 개념으로 접근해서 파생변수를 만든다!

```python
ebola=pd.read_csv('country_timeseries.csv')
```

```python
ebola.columns
```

```python
ebola_pre=pd.melt(ebola, id_vars=['Date','Day'])
```

```python
#str 문법 중에 split함수를 이용해서 나눌 수 있다.
ebola_pre_sp=ebola_pre.variable.str.split('_')
```

```python
ebola_pre_sp
```

```python
ebola_pre['case']= ebola_pre_sp.str.get(0)
ebola_pre['country'] = ebola_pre_sp.str.get(1)
```

```python
ebola_pre
```

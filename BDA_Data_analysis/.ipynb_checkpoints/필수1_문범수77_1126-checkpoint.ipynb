```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.gridspec as gridspec
```

```python
titan = sns.load_dataset('titanic')
```

```python
titan
```

```python
titan.info()
```

```python
titan.isnull().sum()
```

```python
ti_var = ['embarked', 'class', 'deck', 'who']
```

```python
grid = gridspec.GridSpec(2,2) 
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.5, hspace=0.4) 


for index, feature in enumerate(ti_var) : 
    ax = plt.subplot(grid[index])
    
    sns.countplot(x = feature,
                  data = titan, 
                 hue = 'who',
                 palette = 'pastel', 
                  ax = ax)
    plt.xticks(rotation = 90)   
    ax.set_title(f'{feature} Distribution')
    ax.legend().set_visible(False)
```

```python
ti_num = ['pclass', 'sibsp', 'parch', 'survived']
```

```python
grid = gridspec.GridSpec(2,2) 
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.5, hspace=0.4) 

for index, feature in enumerate(ti_num) : 
    ax = plt.subplot(grid[index])
    
    sns.countplot(x = feature,
                  data = titan, 
                 hue = 'survived',
                 palette = 'pastel', 
                  ax = ax)
    plt.xticks(rotation = 90)  
    ax.set_title(f'{feature} Distribution')
    ax.legend().set_visible(False)
```

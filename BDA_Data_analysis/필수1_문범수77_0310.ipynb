### 필수과제1 - weekday 꼭 넣어보기!!

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv('trainby.csv')
train
```

```python
from datetime import datetime
import calendar
```

```python
train['date']=train['datetime'].apply(lambda x: x.split()[0])

train['year']=train['date'].apply(lambda x: x.split('-')[0])
train['month']=train['date'].apply(lambda x: x.split('-')[1])
train['day']=train['date'].apply(lambda x: x.split('-')[2])

train['hour']=train['datetime'].apply(lambda x: x.split()[1].split(':')[0])
train['minute']=train['datetime'].apply(lambda x: x.split()[1].split(':')[1])
train['second']=train['datetime'].apply(lambda x: x.split()[1].split(':')[2])

train
```

```python
train['weekday']=train['date'].apply(lambda date: calendar.day_name[datetime.strptime(date, '%Y-%m-%d').weekday()])
```

```python
drop_features=['casual','registered','datetime','date','month','windspeed']
```

```python
train_fit=train.drop(drop_features, axis=1)
train_fit
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
items=train_fit[['weekday']]

weekday=pd.get_dummies(items, dtype='int')
weekday
```

```python
train_fit_1=pd.concat([train_fit, weekday], axis=1)
train_fit_1=train_fit_1.drop('weekday', axis=1)
train_fit_1
```

```python
## 필요없는 데이터 날리기
train_fit_1=train_fit_1[train_fit_1['weather']!=4]
train_fit_1
```

```python
train_fit_1_X=train_fit_1[['season','holiday','workingday','weather','temp','atemp','weekday_Friday','weekday_Monday','weekday_Saturday','weekday_Sunday','weekday_Thursday','weekday_Tuesday','weekday_Wednesday']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_fit_1_X, train_fit_1['count'], test_size=0.3, random_state=111)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
pred_train=model.predict(X_train)
pred_test=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_train=mean_squared_error(y_true=y_train, y_pred=pred_train)
mse_test=mean_squared_error(y_true=y_test, y_pred=pred_test)
```

```python
print(mse_train)
print(mse_test)
```

```python
train_fit
```

```python
# 일단 위치부터 탐색하고 등차수열 이용해서 x, y 좌표 return
def find_point(num):
    k = 1
    while k * (k + 1) // 2 < num:
        k += 1
    x = num - k * (k - 1) // 2
    y = k + 1 - x
    return x, y

# x, y의 좌표에 해당하는 값 반환
def find_number(x, y):
    return (x + y - 1) * (x + y) // 2 - (y - 1)

# 주어진 p, q 좌표에 대한 연산 수행
def new_operation(p, q):
    x1, y1 = find_point(p)
    x2, y2 = find_point(q)
    x_result = x1 + x2
    y_result = y1 + y2
    return find_number(x_result, y_result)

# 파일에서 입력 읽어오기
with open('input.txt', 'r') as file:
    lines = file.readlines()

# 전체 테스트 케이스 수 읽어오기
T = int(lines[0])

# 각 테스트 케이스에 대해 계산하고 결과 출력
for t in range(1, T + 1):
    p, q = map(int, lines[t].split())
    result = new_operation(p, q)
    print(f"#{t} {result}")
```

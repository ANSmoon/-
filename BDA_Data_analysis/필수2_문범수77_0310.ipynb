### 필수과제2 - target 값 로그스케일링 꼭 진행하기!
- 타깃값을 log 스케일링 진행

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv('trainby.csv')
train
```

```python
from datetime import datetime
import calendar
```

```python
train['date']=train['datetime'].apply(lambda x: x.split()[0])

train['year']=train['date'].apply(lambda x: x.split('-')[0])
train['month']=train['date'].apply(lambda x: x.split('-')[1])
train['day']=train['date'].apply(lambda x: x.split('-')[2])

train['hour']=train['datetime'].apply(lambda x: x.split()[1].split(':')[0])
train['minute']=train['datetime'].apply(lambda x: x.split()[1].split(':')[1])
train['second']=train['datetime'].apply(lambda x: x.split()[1].split(':')[2])

train['weekday']=train['date'].apply(lambda date: calendar.day_name[datetime.strptime(date, '%Y-%m-%d').weekday()])
                                         
train
```

```python
train['count_log']=np.log(train['count'])
train
```

```python
train_fit_X=train[['season','holiday','workingday','weather','temp','atemp']]
train_fit_Y=train['count_log']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_fit_X, train_fit_Y, test_size=0.3, random_state=111)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
pred_train=model.predict(X_train)
pred_test=model.predict(X_test)
```

```python
mse_train=mean_squared_error(y_true=y_train, y_pred=pred_train)
mse_test=mean_squared_error(y_true=y_test, y_pred=pred_test)
```

```python
## mse 출력
print(mse_train)
print(mse_test)
```
